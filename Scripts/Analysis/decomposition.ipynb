```python
cutoff = 3.5
group = 'structures'
kpc_x = 0
kpc_y = 1
deem_file_struct = f'../../Processed_Data/DEEM_10k/Data/{cutoff}/kpca_{group}.hdf5'
deem_kpca_struct = load_structures_from_hdf5(deem_file_struct, datasets=None, concatenate=True)  

iza_file_struct = f'../../Processed_Data/IZA_226onDEEM_10k/Data/{cutoff}/kpca_{group}.hdf5'
iza_kpca_struct = load_structures_from_hdf5(iza_file_struct, datasets=None, concatenate=True)

cod_file_struct = f'../../Processed_Data/COD_196onDEEM_10k/Data/{cutoff}/kpca_{group}.hdf5'
cod_kpca_struct = load_structures_from_hdf5(cod_file_struct, datasets=None, concatenate=True)

fig = plt.figure(figsize=(7.0, 3.5))
axs_struct = fig.add_subplot(1, 2, 1)
axs_env = fig.add_subplot(1, 2, 2)

axs_struct.scatter(deem_kpca_struct[:, kpc_x], deem_kpca_struct[:, kpc_y], alpha=0.05, c=color_list[5], label='DEEM')
axs_struct.scatter(iza_kpca_struct[:, kpc_x], iza_kpca_struct[:, kpc_y], alpha=0.2, c=color_list[1], label='IZA')
axs_struct.scatter(cod_kpca_struct[:, kpc_x], cod_kpca_struct[:, kpc_y], alpha=0.2, c=color_list[2], label='COD')

axs_struct.text(0.95, 0.05, 'Structures', 
                horizontalalignment='right', verticalalignment='bottom',
                transform=axs_struct.transAxes)

axs_struct.set_xlabel(f'KPC {kpc_x+1}')
axs_struct.set_ylabel(f'KPC {kpc_y+1}')

group = 'environments'
deem_file_env = f'../../Processed_Data/DEEM_10k/Data/{cutoff}/kpca_{group}.hdf5'
deem_kpca_env = load_structures_from_hdf5(deem_file_env, datasets=None, concatenate=True)  

iza_file_env = f'../../Processed_Data/IZA_226onDEEM_10k/Data/{cutoff}/kpca_{group}.hdf5'
iza_kpca_env = load_structures_from_hdf5(iza_file_env, datasets=None, concatenate=True)

cod_file_env = f'../../Processed_Data/COD_196onDEEM_10k/Data/{cutoff}/kpca_{group}.hdf5'
cod_kpca_env = load_structures_from_hdf5(cod_file_env, datasets=None, concatenate=True)

axs_env.scatter(deem_kpca_env[:, kpc_x], -deem_kpca_env[:, kpc_y], alpha=0.01, c=color_list[5], label='DEEM')
axs_env.scatter(iza_kpca_env[:, kpc_x], -iza_kpca_env[:, kpc_y], alpha=0.2, c=color_list[1], label='IZA')
axs_env.scatter(cod_kpca_env[:, kpc_x], -cod_kpca_env[:, kpc_y], alpha=0.2, c=color_list[2], label='COD')

axs_env.text(0.95, 0.05, 'Environments', 
                horizontalalignment='right', verticalalignment='bottom',
                transform=axs_env.transAxes)

axs_env.set_xlabel(f'KPC {kpc_x+1}')
axs_env.set_ylabel(f'KPC {kpc_y+1}')

handles = [mpllines.Line2D([0], [0], marker='o', linewidth=0, color=color_list[5]),
           mpllines.Line2D([0], [0], marker='o', linewidth=0, color=color_list[1]),
           mpllines.Line2D([0], [0], marker='o', linewidth=0, color=color_list[2])]
labels = ['DEEM', 'IZA', 'COD']
axs_env.legend(handles=handles, labels=labels, loc='upper left', bbox_to_anchor=(1.0, 1.0))

plt.subplots_adjust(wspace=0.3)

fig.savefig(f'../../Results/{cutoff}/kpca.png', bbox_inches='tight')

plt.show()
```

```python
cantons_struct = np.loadtxt('../../Raw_Data/GULP/IZA_226/cantons.txt', usecols=1, dtype=int)
n_Si_iza = np.loadtxt('../../Processed_Data/IZA_226/n_Si.dat', dtype=int)
cantons_env = np.repeat(cantons_struct, n_Si_iza)
canton_labels = np.unique(cantons_struct)
print(canton_labels)
```

```python
cutoff = 3.5
group = 'structures'
kpc_x = 0
kpc_y = 1
deem_file_struct = f'../../Processed_Data/DEEM_10k/Data/{cutoff}/kpca_{group}.hdf5'
deem_kpca_struct = load_structures_from_hdf5(deem_file_struct, datasets=None, concatenate=True)  

iza_file_struct = f'../../Processed_Data/IZA_226onDEEM_10k/Data/{cutoff}/kpca_{group}.hdf5'
iza_kpca_struct = load_structures_from_hdf5(iza_file_struct, datasets=None, concatenate=True)

fig = plt.figure(figsize=(7.0, 3.5))
axs_struct = fig.add_subplot(1, 2, 1)
axs_env = fig.add_subplot(1, 2, 2)

axs_struct.scatter(deem_kpca_struct[:, kpc_x], deem_kpca_struct[:, kpc_y], alpha=0.05, c=color_list[5], label='DEEM')

for cdx, canton in enumerate(canton_labels):
    canton_idxs = np.nonzero(cantons_struct == canton)[0]
    axs_struct.scatter(iza_kpca_struct[canton_idxs, kpc_x], iza_kpca_struct[canton_idxs, kpc_y], 
                       c=color_list[cdx+1], alpha=0.2, label=f'IZA{canton}')

axs_struct.text(0.95, 0.05, 'Structures', 
                horizontalalignment='right', verticalalignment='bottom',
                transform=axs_struct.transAxes)

axs_struct.set_xlabel(f'KPC {kpc_x+1}')
axs_struct.set_ylabel(f'KPC {kpc_y+1}')

group = 'environments'
deem_file_env = f'../../Processed_Data/DEEM_10k/Data/{cutoff}/kpca_{group}.hdf5'
deem_kpca_env = load_structures_from_hdf5(deem_file_env, datasets=None, concatenate=True)  

iza_file_env = f'../../Processed_Data/IZA_226onDEEM_10k/Data/{cutoff}/kpca_{group}.hdf5'
iza_kpca_env = load_structures_from_hdf5(iza_file_env, datasets=None, concatenate=True)

axs_env.scatter(deem_kpca_env[:, kpc_x], -deem_kpca_env[:, kpc_y], alpha=0.01, c=color_list[5], label='DEEM')

for cdx, canton in enumerate(canton_labels):
    canton_idxs = np.nonzero(cantons_env == canton)[0]
    axs_env.scatter(iza_kpca_env[canton_idxs, kpc_x], -iza_kpca_env[canton_idxs, kpc_y], 
                    c=color_list[cdx+1], alpha=0.2, label=f'IZA{canton}')

axs_env.text(0.95, 0.05, 'Environments', 
                horizontalalignment='right', verticalalignment='bottom',
                transform=axs_env.transAxes)

axs_env.set_xlabel(f'KPC {kpc_x+1}')
axs_env.set_ylabel(f'KPC {kpc_y+1}')

handles = [mpllines.Line2D([0], [0], marker='o', linewidth=0, color=color_list[5]),
           mpllines.Line2D([0], [0], marker='o', linewidth=0, color=color_list[1]),
           mpllines.Line2D([0], [0], marker='o', linewidth=0, color=color_list[2]),
           mpllines.Line2D([0], [0], marker='o', linewidth=0, color=color_list[3]),
           mpllines.Line2D([0], [0], marker='o', linewidth=0, color=color_list[4])]
labels = ['DEEM', 'IZA1', 'IZA2', 'IZA3', 'IZA4']
axs_env.legend(handles=handles, labels=labels, loc='upper left', bbox_to_anchor=(1.0, 1.0))

plt.subplots_adjust(wspace=0.3)

fig.savefig(f'../../Results/{cutoff}/kpca_iza_split.png', bbox_inches='tight')

plt.show()
```
