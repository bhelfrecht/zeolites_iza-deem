```python
# Load FPS-selected DEEM and IZA histograms, 3.5 A SOAP cutoff

distDD = np.loadtxt('../Processed_Data/IZA_OPTonDEEM_1k/Histograms/SOAP/3.5/Linear/FPS/distAA.hist', unpack=True)
distII = np.loadtxt('../Processed_Data/IZA_OPTonDEEM_1k/Histograms/SOAP/3.5/Linear/FPS/distBB.hist', unpack=True)
distDI = np.loadtxt('../Processed_Data/IZA_OPTonDEEM_1k/Histograms/SOAP/3.5/Linear/FPS/distAB.hist', unpack=True)

minDistDD = np.loadtxt('../Processed_Data/IZA_OPTonDEEM_1k/Histograms/SOAP/3.5/Linear/FPS/minDistAA.hist', unpack=True)
minDistDI = np.loadtxt('../Processed_Data/IZA_OPTonDEEM_1k/Histograms/SOAP/3.5/Linear/FPS/minDistAB.hist', unpack=True)
minDistID = np.loadtxt('../Processed_Data/IZA_OPTonDEEM_1k/Histograms/SOAP/3.5/Linear/FPS/minDistBA.hist', unpack=True)

figHist = plt.figure(figsize=(3.54, 3.54))
axsDist = figHist.add_subplot(2, 1, 1)
axsMinDist = figHist.add_subplot(2, 1, 2)

for i in [axsDist, axsMinDist]:
    i.set_ylabel('Frequency')
    
axsMinDist.set_xlabel('Distance')
axsDist.set_ylim([0.0, 15.0])
axsDist.set_xlim([0.001, 1.0])
axsMinDist.set_ylim([0.0, 250.0])
axsMinDist.set_xlim([0.001, 1.0])
axsMinDist.set_xscale('log')
axsDist.set_xscale('log')

axsDist.plot(distDD[0], distDD[1], label='DEEM-DEEM')
axsDist.plot(distII[0], distII[1], label='IZA-IZA')
axsDist.plot(distDI[0], distDI[1], label='DEEM-IZA')
axsDist.fill_between(distDD[0], distDD[1], alpha=0.2)
axsDist.fill_between(distII[0], distII[1], alpha=0.2)
axsDist.fill_between(distDI[0], distDI[1], alpha=0.2)
axsDist.text(0.05, 0.05, u'3.5 \u00c5,\nDistance', horizontalalignment='left', verticalalignment='bottom', 
             transform=axsDist.transAxes, fontsize='medium')

axsMinDist.plot(minDistDD[0], minDistDD[1], label='DEEM-DEEM')
axsMinDist.plot(minDistDI[0], minDistDI[1], label='DEEM-IZA')
axsMinDist.plot(minDistID[0], minDistID[1], label='IZA-DEEM')
axsMinDist.fill_between(minDistDD[0], minDistDD[1], alpha=0.2)
axsMinDist.fill_between(minDistDI[0], minDistDI[1], alpha=0.2)
axsMinDist.fill_between(minDistID[0], minDistID[1], alpha=0.2)
axsMinDist.text(0.05, 0.05, u'3.5 \u00c5, \nMinimum\nDistance', horizontalalignment='left', 
                verticalalignment='bottom', 
                transform=axsMinDist.transAxes, fontsize='medium')


axsDist.legend(frameon=False)
axsMinDist.legend(frameon=False)

figHist.savefig('../Results/DEEM_1k/distances_3.5_linear_FPS.pdf', bbox_inches='tight')
figHist.savefig('../Results/DEEM_1k/distances_3.5_linear_FPS.png', bbox_inches='tight')
plt.show()
```
